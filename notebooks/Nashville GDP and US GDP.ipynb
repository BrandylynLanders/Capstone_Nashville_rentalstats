```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
```

```python
#read in data
df_usgdp = pd.read_csv('C:/Users/brand/Documents/DA8/Projects/capstone_nashville_da8/data/GDP.csv')
df_nashvillegdp = pd.read_csv('C:/Users/brand/Documents/DA8/Projects/capstone_nashville_da8/data/NGMP34980 (06_21).csv')
```

```python
# Plotting line chart for US GDP
plt.plot(df_usgdp['DATE'], df_usgdp['per_change'], label='US GDP')

# Plotting line chart for Nashville GDP
plt.plot(df_nashvillegdp['DATE'], df_nashvillegdp['per_change'], label='Nashville GDP')

# Set the title and axis labels
plt.title('Comparison of US GDP and Nashville GDP')
plt.xlabel('Date')
plt.ylabel('per_change')

# Display the legend
plt.legend()

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
import numpy as np

# Set the width of each bar
bar_width = 0.80

# Set the x-axis values
x = np.arange(len(df_usgdp['DATE']))

# Plotting the stacked bar chart
plt.bar(x, df_usgdp['per_change'], width=bar_width, label='US GDP')
plt.bar(x, df_nashvillegdp['per_change'], width=bar_width, label='Nashville GDP', bottom=df_usgdp['per_change'])

# Set the title and axis labels
plt.title('Comparison of US GDP and Nashville GDP')
plt.xlabel('Date')
plt.ylabel('per_change')

# Set the x-axis tick labels
plt.xticks(x, df_usgdp['DATE'], rotation=45)

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np

# Set the width of each bar
bar_width = 0.80

# Set the x-axis values
x = np.arange(len(df_usgdp['DATE']))

# Define the colors
us_gdp_color = '#7b7aaf'
nashville_gdp_color = '#d1d184'

# Plotting the stacked bar chart with custom colors
plt.bar(x, df_usgdp['per_change'], width=bar_width, label='US GDP', color=us_gdp_color)
plt.bar(x, df_nashvillegdp['per_change'], width=bar_width, label='Nashville GDP', bottom=df_usgdp['per_change'], color=nashville_gdp_color)

# Set the title and axis labels
plt.title('Comparison of US GDP and Nashville GDP')
plt.xlabel('Date')
plt.ylabel('per_change')

# Set the x-axis tick labels
plt.xticks(x, df_usgdp['DATE'], rotation=45)

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
