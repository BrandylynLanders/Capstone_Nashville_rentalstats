```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the interest rates dataset
interest_rates = pd.read_csv('C:/Users/brand/Documents/DA8/Projects/capstone_nashville_da8/data/MORTGAGE30US (1).csv')

# Convert 'DATE' column to datetime
interest_rates['DATE'] = pd.to_datetime(interest_rates['DATE'])

# Extract the year from the 'DATE' column
interest_rates['Year'] = interest_rates['DATE'].dt.year

# Calculate the yearly average of the interest rate
yearly_average = interest_rates.groupby('Year')['MORTGAGE30US'].mean()

# Load the rental price data
rental_prices = pd.read_csv('C:/Users/brand/Documents/DA8/Projects/capstone_nashville_da8/data/avg_monthly_rent.csv')

# Convert 'Date' column to datetime in rental prices data
rental_prices['Date'] = pd.to_datetime(rental_prices['Date'])

# Limit the rental price data to 2016-2023
rental_prices = rental_prices[(rental_prices['Date'].dt.year >= 2016) & (rental_prices['Date'].dt.year <= 2023)]

# Calculate the average rental price per year
avg_rent_per_year = rental_prices.groupby(rental_prices['Date'].dt.year)['Avg Rent'].mean()

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot yearly average interest rates
ax1.bar(yearly_average.index, yearly_average.values)
ax1.set_xlabel('Year')
ax1.set_ylabel('Yearly Average Interest Rate')
ax1.set_title('Yearly Average Interest Rates')

# Plot rental price changes
ax2.plot(avg_rent_per_year.index, avg_rent_per_year.values, color='red', marker='o', linestyle='-')
ax2.set_xlabel('Year')
ax2.set_ylabel('Average Rent')
ax2.set_title('Rental Price Changes')

plt.tight_layout()
plt.show()
```
